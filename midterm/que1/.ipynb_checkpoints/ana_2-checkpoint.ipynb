# Analysis 2
- Displaying all employees with the number of mails they received from top 3 convicts

```python
import os
import re
import time
import nltk
import operator
import datetime
import email.utils
import numpy as np
from email.parser import Parser
from matplotlib import pyplot as plt
```

```python
all_employees = ['lay-k', 'skilling-j', 'whalley-g']
emp_msg_list = []
emp_name_list = []

lay_emailid_list = []
skilling_emailid_list = []
whalley_emailid_list = []

for emp in all_employees:
    filepath_emp = '~\\midterm\\data\\enron\\maildir\\' + emp 
    filepath_actual = os.path.expanduser(filepath_emp)
    emp_name_list.append([emp,filepath_actual])   

for emp, emp_path in emp_name_list:
    for subdir, dirs, files in os.walk(emp_path):
        for file in files:
            if(file[0] != '.'):
                try:
                    response = open(os.path.join(subdir, file), 'r', errors = 'ignore')
                    email_emp = Parser().parsestr(response.read())
        
                    if emp == 'lay-k':                   
                        match = re.findall(r'[\w\.-]+@[\w\.-]+', email_emp['To'])
                        for x in match:
                            lay_emailid_list.append(x)
                    
                    elif emp == 'skilling-j':                    
                        match = re.findall(r'[\w\.-]+@[\w\.-]+', email_emp['To'])
                        for x in match:
                            skilling_emailid_list.append(x)
                    
                    elif emp == 'whalley-g':                    
                        match = re.findall(r'[\w\.-]+@[\w\.-]+', email_emp['To'])
                        for x in match:
                            whalley_emailid_list.append(x)
                except(RuntimeError, TypeError, NameError):
                    pass
```

```python
lay_emailid_unique_dict = set(lay_emailid_list)
skilling_emailid_unique_dict = set(skilling_emailid_list)
whalley_emailid_unique_dict = set(whalley_emailid_list)

count_lay = 0
lay_dict = {}
for x in lay_emailid_unique_dict:
    for y in lay_emailid_list:
        if x == y:
            count_lay = count_lay + 1
    lay_dict.update({x:count_lay})
    count_lay = 0
    
count_skilling = 0
skilling_dict = {}
for x in skilling_emailid_unique_dict:
    for y in skilling_emailid_list:
        if x == y:
            count_skilling = count_skilling + 1
    skilling_dict.update({x:count_skilling})
    count_skilling = 0
    
count_whalley = 0
whalley_dict = {}
for x in whalley_emailid_unique_dict:
    for y in whalley_emailid_list:
        if x == y:
            count_whalley = count_whalley + 1
    whalley_dict.update({x:count_whalley})
    count_whalley = 0
```

```python
sorted_lay_dict = sorted(lay_dict.items(), key=operator.itemgetter(1), reverse = True)
sorted_skilling_dict = sorted(skilling_dict.items(), key=operator.itemgetter(1), reverse = True)
sorted_whalley_dict = sorted(whalley_dict.items(), key=operator.itemgetter(1), reverse = True)
```

```python
y_values_lay = [x[1] for x in sorted_lay_dict[:7] if x[0] not in ['kenneth.lay@enron.com', 'klay@enron.com']]
x_values_lay = [y[0] for y in sorted_lay_dict[:7] if y[0] not in ['kenneth.lay@enron.com', 'klay@enron.com']]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title("Kenneth Lay Mail Count")

width = .35
ind = np.arange(len(y_values_lay))
plt.bar(ind, y_values_lay, width=width)
plt.xticks(ind + width / 2, x_values_lay)

fig.autofmt_xdate()

save_path = os.path.expanduser('~\\midterm\\que1\\ana_2\\Kenneth Lay Mail Count.pdf')

plt.savefig(save_path)

plt.show()
```
